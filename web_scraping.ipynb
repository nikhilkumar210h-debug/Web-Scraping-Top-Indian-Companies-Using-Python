```python
data =[]
# row ={}
df = pd.DataFrame() 
```

```python
for x in range(1,501):
 headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
 url ='https://www.ambitionbox.com/list-of-companies?page={}'.format(x)
 web_page = requests.get(url,headers=headers).text
 soup=BeautifulSoup(web_page,'lxml')
 for a in soup.find_all('div',class_='companyCardWrapper'):
    row ={}
    tit = a.find('h2').get_text(strip=True)
    about = a.find('span', class_="companyCardWrapper__interLinking").get_text(strip=True)
    rt_tag = a.find('span', class_="companyCardWrapper__companyRatingCount")
    rt = rt_tag.get_text(strip=True) if rt_tag else None
    row['About'] = about
    row['Rating'] = rt
    row['Company'] = tit
    # data.append(row)
    # print(about)
    for i in a.find_all('a',class_='companyCardWrapper__ActionWrapper'):
        title = i.find('span', class_="companyCardWrapper__ActionTitle").get_text(strip=True)
        count = i.find('span', class_="companyCardWrapper__ActionCount").get_text(strip=True)
        row[title] = count
        # print(count)
    data.append(row)
        
        
    
df = pd.DataFrame(data)

import re

def convert_to_float(val):
    if isinstance(val, str):
        val = val.strip().lower()
        val = val.replace('(', '').replace(')', '')   
        val = val.replace(',', '')                    
        if 'l' in val:   # lakh
            return float(val.replace('l','')) * 100000
        elif 'k' in val: # thousand
            return float(val.replace('k','')) * 1000
        else:
            try:
                return float(val)
            except:
                return None
    return val
    
df['Rating'] = df['Rating'].apply(convert_to_float)
df['Salaries'] = df['Salaries'].apply(convert_to_float)
df['Reviews'] = df['Reviews'].apply(convert_to_float)
df['Jobs'] = df['Jobs'].apply(convert_to_float)
df['Benefits'] = df['Benefits'].apply(convert_to_float)
df['Interviews'] = df['Interviews'].apply(convert_to_float)
df['Photos'] = df['Photos'].apply(convert_to_float)
df = df.sort_values(by='Salaries', ascending=False,ignore_index=True)
# print(df.head())
```
